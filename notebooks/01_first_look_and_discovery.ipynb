## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### Redo Discovery A for pages with `impressions_90d > 0`

```python
visible_impressions_df = df[df["impressions_90d"] > 0]
corr_filtered = visible_impressions_df["search_volume"].corr(visible_impressions_df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d (for impressions > 0): {corr_filtered:.3f}")
print(f"Original correlation (all pages): {corr:.3f}")

if abs(corr_filtered) > abs(corr):
    print("\nThe correlation increases slightly when considering only pages with impressions, but it remains very low.")
else:
    print("\nThe correlation remains very low, even when considering only pages with impressions. It implies that keyword search volume still does not strongly predict actual traffic, even for visible pages.")
```

### Discovery B — Which `content_type` has the worst CTR *at the same position tier*?

```python
ctr_by_pos_ctype = visible.groupby(["position_tier", "content_type"])["ctr"].mean().unstack()

# Identify the content_type with the lowest CTR for each position_tier
worst_ctr_ctype_per_pos = ctr_by_pos_ctype.idxmin(axis=1)
min_ctr_per_pos = ctr_by_pos_ctype.min(axis=1)

print("Content type with the worst CTR for each position tier:")
for pos_tier, worst_ctype in worst_ctr_ctype_per_pos.items():
    print(f"  - {pos_tier}: {worst_ctype} (min CTR: {min_ctr_per_pos[pos_tier]:.4f})")

print("\nThis analysis highlights content types that might be underperforming within specific search result positions. For instance, in 'page_1' positions, 'blog article' has the lowest average CTR, suggesting it's less effective at capturing clicks compared to other content types, even when ranking highly.")
```

### Find another belief to test — Does `content_age_days` relate to `trend_direction`?

```python
content_age_by_trend = df.groupby("trend_direction")["content_age_days"].median().sort_values(ascending=False)
print("Median content age (days) by trend direction:")
print(content_age_by_trend.round(0).to_string())

print("\nThis shows how the median age of content varies based on its trend direction. \nInterestingly, 'down' trending content appears to be younger on average (216 days) compared to 'up' (292 days) or 'stable' (300 days) content. This suggests that newer content might be experiencing a decline, while older content is more stable or improving.")
```

### Find another belief to test — Does `avg_position` relate to `ctr`?

```python
corr_pos_ctr = visible["avg_position"].corr(visible["ctr"])
print(f"Correlation between avg_position and ctr (for impressions >= 100): {corr_pos_ctr:.3f}")

print("\nThis negative correlation indicates that as the average position gets lower (i.e., closer to position 1), the CTR tends to increase. This is an expected and strong relationship in search results: better positions generally lead to higher click-through rates.")
```
