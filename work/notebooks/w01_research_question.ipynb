## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

After analyzing the initial dataset slice, we found that:

1. Out of 30,000 pages, the baseline declining rate is 54.2% (0.542), meaning over half the pages are actively losing visibility.

2. Traditional metrics are misleading: the correlation between a keyword's estimated search_volume and actual impressions_90d is practically zero (0.001), proving we need a smarter ML model instead of relying on simple search volume metrics.

3. The median word count for declining pages (down trend) is 2909 words compared to growing pages (up trend) at 2848 words, proving that simply writing longer text is not the solution to fix declining traffic.

```python
import pandas as pd
import os

# Scan the Google Colab directories to locate the dataset file automatically
csv_path = None
for root, dirs, files in os.walk("."):
    if "content_refresh_anonymized.csv" in files:
        csv_path = os.path.join(root, "content_refresh_anonymized.csv")
        break

# If the file is found locally, load it; otherwise, use the safe remote URL fallback
if csv_path:
    print(f"File found at: {csv_path}")
    df = pd.read_csv(csv_path)
else:
    print("Local file not found, loading directly from the source repository...")
    url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
    df = pd.read_csv(url)

# 1. Calculate the overall baseline declining rate
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)
declining_rate = df["is_declining_label"].mean()

# 2. Compute the baseline correlation between search volume and actual traffic
corr = df["search_volume"].corr(df["impressions_90d"])

# 3. Calculate median word counts grouped by trend direction
wc_medians = df.groupby("trend_direction")["word_count"].median()

# Print out the verified metrics
print(f"\nTotal Pages Analysed: {df.shape[0]}")
print(f"1. Overall Declining Rate: {declining_rate:.3f}")
print(f"2. Search Volume vs Actual Impressions Correlation: {corr:.3f}")
print("\n3. Median Word Counts by Trend:")
print(wc_medians.to_string())
```
