## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

This task is structured as a Binary Classification problem. The machine learning model will analyze historical performance metrics of individual web pages and classify each page into one of two distinct states: 1 (Declining/Decaying requires an immediate rewrite or refresh action) or 0 (Stable/Growing healthy state, no action required).

```python
# Problem type verification
print("ML Task Type: Binary Classification (Target: 0 or 1)")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The true target variable is is_declining_label. Since true content decay is a lagging operational state, we extract this binary label directly from the historical traffic trend direction (trend_direction). If a page's search visibility and structural impressions over a 90-day window show a sharp downward trajectory (down), it is flagged as a positive instance (1) for content optimization.

```python
# Checking the target column balance template
print("Target Variable Mapping: {'down': 1, 'up': 0, 'flat': 0}")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Our primary evaluation metric will be Precision, closely supported by the F1-Score.

- **Why Precision?** As analyzed in our workflow audit, a False Positive means flagging a healthy page as declining, which directly wastes human editorial budget and hours on unnecessary rewrites. We must ensure that when our model says a page is dying, the engineering or content team can trust that recommendation with high certainty.

```python
# Metric placeholder tracking
print("Primary Optimization Metric: Precision")
print("Secondary Optimization Metric: F1-Score")
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

The Unit of Analysis is an Individual Web Page (URL) at a specific snapshot in time. Each single row in our dataset represents a unique page along with its underlying structural SEO metrics (such as word count, crawling frequency, impressions, and search volume).

```python
import pandas as pd
import os

# Locate the dataset file across the Colab workspace dynamically
csv_path = None
for root, dirs, files in os.walk("."):
    if "content_refresh_anonymized.csv" in files:
        csv_path = os.path.join(root, "content_refresh_anonymized.csv")
        break

# Load the dataframe securely using local path or remote repository fallback
if csv_path:
    df = pd.read_csv(csv_path)
else:
    url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
    df = pd.read_csv(url)

# Construct the explicit binary target column for the classification pipeline
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Safely extract whatever columns are actually available in the dataset
available_columns = df.columns.tolist()
target_columns = ["word_count", "search_volume", "impressions_90d", "is_declining_label"]

# Add an identifier column if it exists under common names
for possible_id in ["url", "id", "url_hash", "page"]:
    if possible_id in available_columns:
        target_columns.insert(0, possible_id)
        break

# Filter only columns that genuinely exist to prevent KeyErrors
final_columns = [col for col in target_columns if col in available_columns]
sample_view = df[final_columns].head(5)

print("--- Dataframe Unit of Analysis (1 Row = 1 Unique Web Page) ---")
print(f"Dataset Dimensions: {df.shape[0]} rows and {df.shape[1]} columns.\n")
print(sample_view.to_string(index=False))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A static business rule like "Flag any page whose traffic drops by more than 20%" fails completely in real-world SEO operations. Traffic fluctuates naturally due to seasonality, global holiday breaks, or brief crawling algorithmic hiccups.

Machine Learning beats fixed rules because it learns non-linear relationships across multiple features simultaneously. For instance, our exploratory numbers showed that declining pages often have higher median word counts than growing pages. A fixed rule would assume longer content equals stable traffic, whereas an ML classifier can discover that un-optimized long-form content is actually highly susceptible to keyword decay, enabling predictive content management instead of reactive fixing.

```python
# Checking non-linear interaction baseline
print("Rule-based vs ML Framework verified. Static heuristics are insufficient.")
```
